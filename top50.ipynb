```python
MUSIXMATCH_APIKEY= 'XXXX'
import pandas as pd
from pprint import pprint
top50 = pd.read_csv("top50.csv", engine="python", encoding='latin1')
```

```python
top50['song_name'] = top50['Track.Name'].str.lower().str.replace('\(.*\)', '').str.replace(' ', '-').str.replace('-$', '').str.replace("'", "")
top50['artist_name'] = top50['Artist.Name'].str.lower().str.replace(' ', '-')
top50['merge_index'] = (top50['artist_name'] + '_' + top50['song_name'])
```

```python
import requests

def get_lyric(merge_index):
    pprint(merge_index)
    artist = merge_index.split('_')[0].replace("-", " ")
    song_name = merge_index.split('_')[1].replace("-", " ")
    try: 
        r = requests.get("https://api.musixmatch.com/ws/1.1/track.search?apikey=%s&q_artist=%s&q_track=%s" % (MUSIXMATCH_APIKEY, artist, song_name))
        print(r.status_code)
        #pprint(r.json())
        if r.status_code == requests.codes.ok:
            l = requests.get("https://api.musixmatch.com/ws/1.1/track.lyrics.get?apikey=%s&track_id=%s" % (MUSIXMATCH_APIKEY, r.json()['message']['body']['track_list'][0]['track']['track_id']))
            if l.status_code == requests.codes.ok:
                    return l.json()['message']['body']['lyrics']['lyrics_body']
    except:
        return "ERROR"
    
top50['song_lyric'] = top50['merge_index'].map(get_lyric)

        
```

```python
top50.to_csv("top50_lyrics.csv")
```
